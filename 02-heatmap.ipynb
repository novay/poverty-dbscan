```python
# %pip install folium
import pandas as pd
import folium
from folium.plugins import HeatMap

df = pd.read_csv('./datasets/survey-preprocessed.csv')
df.sample(5)
```

```python
def generateBaseMap(default_location = [df['latitude'].median(), 
                                        df['longitude'].median()], 
                    default_zoom_start = 11):
    
    return folium.Map(location = default_location, 
                      control_scale = True, 
                      zoom_start = default_zoom_start)
```

```python
df['Count'] = 1
base_map = generateBaseMap()

HeatMap(data = df[['latitude', 'longitude', 'Count']]
            .groupby(['latitude', 'longitude'])
            .sum()
            .reset_index()
            .values.tolist(),
        radius = 7,
        max_zoom = 18).add_to(base_map)

display(base_map)
```
