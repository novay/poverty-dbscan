### **1. Library**

```python
import pandas as pd
pd.set_option('display.max_columns', None)

import seaborn as sns
sns.set_palette('Set2')
sns.set_style(style = 'darkgrid')
import matplotlib.pyplot as plt
```

### **2. Read Dataset (Filtered)**

```python
df = pd.read_csv('./datasets/survey-preprocessed.csv')
df.shape
```

**2.1 (Masuk) vs (Tidak Masuk) KRITERIA MISKIN**

```python
df['kecamatan'] = df['kecamatan'].str.upper()

fig, ax = plt.subplots()
sns.countplot(y='kecamatan', hue='status', data=df, palette={0: 'red', 1: 'green'})

ax.set_title('Kriteria Penerima Bantuan')
ax.set_xlabel('Jumlah')
ax.set_ylabel('Kecamatan')

legend = plt.legend(title='Status')

legend_labels = {0: 'Tidak Masuk Kriteria', 1: 'Masuk Kriteria'}
for text, label in zip(legend.get_texts(), legend_labels.values()):
    text.set_text(label)

fig.set_size_inches(10, 10)
plt.show()
```

Diatas merupakan distribusi perbandingan dua jenis hasil dari survey yang dilakukan yaitu:
- Memenuhi Kriteria, dan;
- Yang tidak memenuhi kriteria.
