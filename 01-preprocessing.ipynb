```python
import pandas as pd

df = pd.read_csv('datasets/survey.csv', sep=";")
df.head()
```

```python
# Hapus data yang bernilai null pada titik koordinat
df = df[pd.notnull(df['latitude'])]
df = df[pd.notnull(df['longitude'])]

# Ganti koma (',') dengan titik ('.') dalam kolom 'latitude', 
# Lalu konversi ke tipe float
df['latitude'] = df['latitude'].str.replace(',', '.').astype(float)

# Ekstraksi hanya nilai numeric, untuk menghindari karakter khusus seperti Â° (derajat)
df['longitude'] = pd.to_numeric(df['longitude'].str.extract('([-+]?\d*\.\d+|\d+)', expand=False), errors='coerce')
```

```python
df = df.drop(['uuid', 'group_id', 'rt', 'alamat'], axis=1)

# Reset the index of the sorted DataFrame
df = df.reset_index(drop=True)
df.to_csv('datasets/survey-preprocessed.csv')

print('Total Data: ' + str(len(df)))
print('======================')
df.head()
```
